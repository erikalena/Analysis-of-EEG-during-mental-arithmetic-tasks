# Class masks visualization

Class-specific essential frequency masks can be trained for e given classification problem (MS, CQ) using the python script `get_class_mask.py`. 

Once these masks are available further analysis can be performed to understand the characteristics of the masks, in particular which frequency bands are most important for each class and which brain areas (electrodes) are the most useful for the classification.

In this notebook, it is shown how to use functions available in `` python file to visualize the class masks and to extract the most important frequency bands and electrodes for each class.

```python
import os
os.chdir('../src_code')
import numpy as np

from utils.read_data import CHANNEL_NAMES
from utils.plot_masks_results import plot_mask_sum, plot_topographic_mask, plot_topographic_mask_bands
```

## Mental state (MS) classification

For each class a sum of the masks is calculated to obtain a single mask that represents the most important frequency bands for the classification of that class. The masks are then visualized using a topographic plot.


```python
# provide the path to the results folder
classification_type = 'ms'
RESULTS_FOLDER = f'./results_class_mask/results_class_mask_{classification_type}_second_1/' # provide path to the results folder

path_class_0 = f'{RESULTS_FOLDER}/class_0'
path_class_1 = f'{RESULTS_FOLDER}/class_1' 
plot_mask_sum(path_class_0, path_class_1, CHANNEL_NAMES, normalization=False)
```

### Topographic plot of the class masks

The topographic plots show instead which are the most used electrodes for the classification of each class. 

```python
path_0 = './results_class_mask/results_class_mask_ms_first_1/class_0/'
path_1 = './results_class_mask/results_class_mask_ms_first_1/class_1/'

subject = 1
edf_file = f'../eeg_data/Subject{subject:02}_2.edf'

plot_topographic_mask(path_0, path_1, edf_file, CHANNEL_NAMES, normalization=False)
```

```python
path_0 = './results_class_mask/results_class_mask_ms_first_2/class_0/'
path_1 = './results_class_mask/results_class_mask_ms_first_2/class_1/'


plot_topographic_mask(path_0, path_1, edf_file, CHANNEL_NAMES, normalization=False)
```

We can also visualize the topographic map for each frequency band, trying to understand the information coming from which electrodes is used the most for each frequency band.

```python
path_0 = './results_class_mask/results_class_mask_ms_first_1/class_0/'
path_1 = './results_class_mask/results_class_mask_ms_first_1/class_1/'

plot_topographic_mask_bands(path_0, path_1, edf_file, CHANNEL_NAMES, normalization=False)
```
